```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# autsl
total_frames = 1731583+261484+220573
lh_missed = 232473+34524+38811
lh_not_missed = total_frames-lh_missed
print("Percentager of LH missed = " + str((lh_missed/total_frames)*100))
rh_missed = 182998 + 25677 + 20344
rh_not_missed = total_frames-rh_missed
print("Percentager of RH missed = " + str((rh_missed/total_frames)*100))
pose_missed = 1071 + 14 + 32
pose_not_missed = total_frames-pose_missed
print("Percentager of pose missed = " + str((pose_missed/total_frames)*100))
```

```python
# plot missed frames
labels = ["Left-hand", "Right-hand", "Body-Pose"]
x = np.arange(len(labels))
width=0.35

missed_detections = [lh_missed, rh_missed, pose_missed]
detections = [lh_not_missed, rh_not_missed, pose_not_missed]

fig, ax = plt.subplots()
rects1 = ax.bar(x-width/2, missed_detections, width, label='Missed Frame Detections')
rects2 = ax.bar(x+width/2, detections, width, label='Successful Frame Detections')

ax.set_ylabel('Number of Frames', fontsize=16)
ax.set_title('MediaPipe Frame Dections for AUTSL.', fontsize=20)
ax.set_xticks(x, labels, fontsize=16, rotation=0)

ax.legend(['Missed Frame Detections', 'Successful Frame Detections'], fontsize=11)
fig.tight_layout()
# plt.show()
plt.savefig('MP_detections_autsl.png')
```
